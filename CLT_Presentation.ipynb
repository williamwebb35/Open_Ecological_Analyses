```python
pivoted_df_corvids = df_corvids.pivot(index='cal_yr', columns='com_name', values='how_many')
pivoted_df_corvids
```

```python
pivoted_df_corvids= pivoted_df_corvids.fillna(0)
pivoted_df_corvids
```

```python
flat_corvids_df = pivoted_df_corvids.reset_index()
flat_corvids_df
```

```python
# 1. Create data
x = flat_corvids_df['cal_yr']
y1 = flat_corvids_df['American Crow'] #crows
y2 = flat_corvids_df['Common Raven'] #ravens
y3 = flat_corvids_df["Steller's Jay"]# st jays
y4 = flat_corvids_df['Western Scrub-Jay']# scrub jays

# 2. Plot the lines
plt.plot(x, y1, color = 'red',label='Crows')
plt.plot(x, y2, color = 'black', label='Ravens')
plt.plot(x, y3, color = 'green', label="Steller's Jays")
plt.plot(x, y4, color = 'orange', label='Western Scrub-Jay')

# 3. Add labels, title, and a legend for clarity
plt.xlabel('Year')
plt.ylabel('Number Counted')
plt.title('Corvid Abundance',fontsize = 22)
plt.legend() # This displays the labels added in plt.plot()

plt.savefig('Corvid_Abundance.png') # Saves as a PNG file
```
